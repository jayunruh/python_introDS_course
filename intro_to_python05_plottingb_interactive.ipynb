```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
import webbrowser
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
df = pd.read_csv('https://www.dropbox.com/s/czjccph6bto67ge/Compositions.csv?dl=1')
df = df.fillna(1)
df['SqrtViews'] = np.power(df['Views'],.5)
```

```python
df
```

```python
f=go.FigureWidget(
    px.bar(df, x='Year Completed', y='SqrtViews', color='Type', hover_data=['Composition', 'Link'], height=600, title='Popularity of Beethoven Compositions')
    )

def click_fn(trace, points, state):
    
    if (len(points.point_inds)>0):
        link = (f.data[points.trace_index]['customdata'][points.point_inds[-1]])[1]
        print(link)
        if (isinstance(link, str)):
            webbrowser.open(link) 

for a in f.data:
    a.on_click(click_fn)

f
```
